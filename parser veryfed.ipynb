```python
tokenlist = 'https://raw.githubusercontent.com/wavesplatform/wavesplatform.core.js/master/src/core/waves.money.js'
```

```python
import urllib2
skip = 0 

assets = {}

def clearstr(line):
    line = line.split(':')[1].replace(',','').replace('\'','').replace(' ','').replace('\n','')
#     print line
    return line


for line in urllib2.urlopen(tokenlist):
#     print line
    skip = skip + 1
    if skip < 50: continue
    if 'invalidateCache' in line: break
        
    if "id" in line:
        idt = clearstr(line)
        assets[idt] = {'id':idt}
    if "displayName" in line:
        assets[idt].update({'displayName':clearstr(line)})
    if "shortName" in line:
        assets[idt].update({'shortName':clearstr(line)})
    if "precision" in line:
        assets[idt].update({'precision':clearstr(line)})
    if "verified" in line:
        assets[idt].update({'verified':clearstr(line)})
    
```

```python
assets
```

```python
import pandas

atable = pandas.DataFrame.from_dict(assets, orient='index').reset_index()[['id','displayName','shortName','precision','verified']]
atable
```

```python
atable.to_csv('verified.csv')
```
